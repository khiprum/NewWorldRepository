```
import numpy as np
import math
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from matplotlib import cm

fig = plt.figure()
X = np.arange(0, 5, 0.1)
xlen = len(X)
Y = np.arange(0, 5, 0.1)
ylen = len(Y)
X, Y = np.meshgrid(X, Y)
R = np.sqrt(np.log(X)**2 + np.log(Y)**2)
Z = np.exp(R)
color_line = np.zeros((10, 3))
color_line[:, 1:] = 0.1
color_line[:, 0] = np.linspace(0, 1, 10)
curves = plt.contour(X, Y, Z, np.linspace(0, 5, 10),colors=color_line)
plt.clabel(curves)
plt.axes().set_aspect(1)
plt.title(r'$z=exp(sqrt(ln(x)**2+ln(y)**2))$', fontsize=12)
plt.show()
fig.savefig('квадрат логарифма.jpg', quality = 100)
```
